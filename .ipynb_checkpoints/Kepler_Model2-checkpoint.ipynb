```python
# import dependencies 
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
```

## Read & Preview the Data, Perform Basic Cleaning 

```python
kepler = pd.read_csv("Resources/cumulative.csv")
print(kepler.shape)
kepler.head()
```

```python
kepler =  kepler.dropna(axis="columns", how="all")
kepler = kepler.dropna()
print(kepler.shape)
```

## Select features

```python
kepler.columns
```

```python
X = kepler[['koi_score', 'koi_fpflag_nt', 'koi_fpflag_ss','koi_fpflag_co', 'koi_fpflag_ec', 
               'koi_period','koi_time0bk','koi_impact','koi_duration','koi_depth','koi_prad','koi_teq', 'koi_insol','koi_model_snr', 'koi_tce_plnt_num','koi_steff','koi_slogg','koi_srad', 'ra', 'dec', 'koi_kepmag']]
X.head()
```

## Train Test Split data

```python
y=kepler["koi_pdisposition"]
print(y.shape)
print(X.shape)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
X_train.head()
```

## Pre-processing
scale the data using the MinMaxScaler and perform some feature selection 

```python
from sklearn.preprocessing import MinMaxScaler

X_scaler = MinMaxScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Create the model - SVC Model

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
model = SVC(kernel="linear")
model
```

### Fit the model with training data

```python
model.fit(X_train_scaled, y_train)
```

### Score model using training & testing data

```python
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

### Generate classification report using test data

```python
predictions = model.predict(X_test_scaled)
print(classification_report(y_test, predictions,
                            target_names=["CANDIDATE","FALSE POSITIVE"]))
```

## Hyperparamter Tuning
Tuning the model's parameters using GridSearchCV

```python
# create the GridSearchCV model
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.001, 0.01, 1, 5, 10, 50],
              'gamma': [0.0001, 0.0005, 0.001, 0.005, 0.05, 0.5]}

grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
# train the model with GridSearch
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
